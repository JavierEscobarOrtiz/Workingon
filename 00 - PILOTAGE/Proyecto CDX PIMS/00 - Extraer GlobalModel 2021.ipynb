```python
import pandas as pd
import numpy as np
import shutil
from zipfile import ZipFile
import os
from datetime import datetime
```

```python
# Busco los meses disponibles en el directorio
ruta_directorio_st = r"\\cepnas\planificacion\PIMS PL\Advanced Optimization\2021"
lista = os.listdir(ruta_directorio_st)
meses_names = [x for x in lista if x.startswith("0") or x.startswith("1")]
# Creo las ubicaciones
meses_ruta = []
for mes in range(len(meses_names)):
    meses_ruta.append(ruta_directorio_st + "\\" + meses_names[mes] + "\\")
    
print(meses_names)
print(meses_ruta)
```

```python
# mes = meses_names[0]
# mes_ruta = meses_ruta[0]

def encontrar_zips(mes, mes_ruta):
    """Genera un df con los zips en las diferentes carpetas,
    devuelve NaN si no encuentra un .zip"""
    
    # Recorro mes seleccionado
    mes = mes
    mes_ruta = mes_ruta
    semanas_bruto = os.listdir(mes_ruta)
    semanas = []
    
    # Elimino si hay algun archivo zip
    for archivo in semanas_bruto:
        # Devuelve True si es carpeta, si es archivo False
        if os.path.isdir(mes_ruta + "\\" + archivo):
            semanas.append(archivo)
    
    semanas_names = []
    # lista con nombres de versiones 
    for s in range(len(semanas)):
        semanas_names.append(semanas[s].replace(" ", "_"))

    # lista con las ubicaciones
    semanas_ruta = []
    for s in semanas:
        semanas_ruta.append(mes_ruta + "\\" + s + "\\")

    # Encuentro los ficheros a extraer
    archivos_zip_names = []
    archivos_zip_ruta = []
    for s in range(len(semanas_ruta)):
        archivos = os.listdir(semanas_ruta[s])
        for archivo in archivos:
            if archivo.endswith(".zip"):
                archivo_zip = archivo
                break
            else:
                # Si no hay un archivo .zip pongo un NaN
                archivo_zip = "NaN"
        archivos_zip_names.append(archivo_zip)
        archivos_zip_ruta.append(semanas_ruta[s] + "\\" + archivo_zip)
    
    
    df_mes = pd.DataFrame({"Mes":mes, "Semana":semanas_names, "Archivos Zip":archivos_zip_names, "Rutas Archivos Zip":archivos_zip_ruta})
    df_mes["Mes"] = df_mes["Mes"].str.replace(" ", "_")
    
    # Si el archivo no es un zip pongo un Nan
    if not df_mes.empty:
        df_mes["Archivos Zip"].loc == df_mes["Archivos Zip"].str.endswith(".zip")
    
    print("Datos cargados de " + mes)
    
    return df_mes
```

```python
# Llamo a la funcion
año = "2021"
df_meses = pd.DataFrame(columns=["Mes", "Semana", "Archivos Zip", "Rutas Archivos Zip"])

for i in range(len(meses_names)):
    df_mes = encontrar_zips(meses_names[i], meses_ruta[i])
    df_meses = pd.concat([df_meses, df_mes]).reset_index(drop = True)

pd.set_option('display.max_colwidth', None)
display(df_meses)
pd.reset_option('display.max_colwidth')
```

```python
año = "2021"
destino_extraccion = r'C:\Users\JESCOBARO\Desktop\CDX_PIMS\Extracciones'
df = df_meses.copy()
mes = df["Mes"].tolist()
semanas_names = df["Semana"].tolist()
archivos_zip = df["Archivos Zip"].tolist()
archivos_zip_ruta = df["Rutas Archivos Zip"].tolist()

print(datetime.now().strftime('%d-%m-%Y %H:%M:%S') + ": Comenzando el proceso...")
for archivo in archivos_zip_ruta:
       
    # Encuentro su indice
    i = archivos_zip_ruta.index(archivo)
    
    #Calculo progreso
    total = len(archivos_zip_ruta)
    progreso = round(100 *  (i+1)/total)
    
    if archivos_zip[i] == "NaN":
        # No tengo archivo, no lo copio
        pass
    else:
        # Create a ZipFile Object and load sample.zip in it
        with ZipFile(archivo, 'r') as zipObj:
           # Get a list of all archived file names from the zip
            listOfFileNames = zipObj.namelist()
            # Me quedo solo con la carpeta Global Model dentro de los zips
            carpeta = "Global Model"
            filtrado = [x for x in listOfFileNames if carpeta in x]
            # Iterate over the file names
            for fileName in filtrado:
                # Check filename endswith Mdb
                if fileName.endswith('.Mdb'):
                    # Extract a single file from zip
                    fichero = semanas_names[i]
                    nombre_access = fileName
                    zipObj.extract(fileName, destino_extraccion)    

        if nombre_access is not None:
            original = destino_extraccion + "\\" + nombre_access
            print("Extrayendo " + mes[i] + " versión " + fichero + ", " + str(progreso) + "% completado")
            target = r'C:\Users\JESCOBARO\Desktop\CDX_PIMS\PIMS\\' +  mes[i] + "_" + año + "-" + fichero + "_Results.Mdb" 
            print("Guardado en... " + target)

        shutil.copyfile(original, target)

print(datetime.now().strftime('%d-%m-%Y %H:%M:%S') + ": Proceso finalizado")
```

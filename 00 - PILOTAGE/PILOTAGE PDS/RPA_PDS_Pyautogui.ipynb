```python
# Genero lista con dfs incluyendo los pasos 3 correspondientes

pasos_3 = df_web_accion.pasos.to_list()
lista_df = []

for paso in pasos_3:
    
    # Agrego el paso 3 correspondiente, ordeno y genero el df
    df_aux = df_web_accion[df_web_accion["pasos"] == paso]
    df_aux = pd.concat([df_web_tipo, df_aux])
    df_aux["ordenar"] = df_aux["pasos"].str[:2].astype(int)
    df_aux = df_aux.sort_values(by="ordenar", ascending = True).drop(columns = ['ordenar']).reset_index(drop = True)
    
    # Genero la variable df_paso
    globals()["df_" + paso] = df_aux.copy()
    globals()["df_" + paso].name = f'df_{paso}'
    lista_df.append(globals()["df_" + paso])
    
print([x.name for x in lista_df])
```

```python
# TIEMPO TOTAL 22 minutos aprox

# Para quitar algun df ya ejecutado incluirlo como str en la lista
no_ejecutar = []

start_time_accion_total = time.time()

for df in lista_df:
    if df.name not in no_ejecutar:
        print("Ejecutando acciones de:", df.name)
        print("************************************")
        rpa_pds.ejecutar_rpa(df)
        
        
end_time_accion_total = time.time()
total_time_accion = end_time_accion_total - start_time_accion_total
print("====================================")
print("Total time:", time.strftime("%H:%M:%S", time.gmtime(total_time_accion))) 
```

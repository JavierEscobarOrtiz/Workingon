```python
import pandas as pd
```

```python
titulos = pd.DataFrame({"Episodio" : [1,3,4,5,6,7,9],
                          "Titulo" : ["La amenaza fantasma", "La venganza de los Sith", 
                                      "Una nueva esperanza", "El imperio contraataca",
                                      "El retorno del Jedi", "El despertar de la Fuerza",
                                      "El ascenso de Skywalker"]
                       })

duracion = pd.DataFrame({"Episodio" : [1,2,3,4,7,8],
                         "Duracion (min)" : [136, 142, 140, 121, 138, 152]})

duracion_2 = pd.DataFrame({"Numero" : [1,2,3,4,7,8],
                         "Duracion (min)" : [136, 142, 140, 121, 138, 152]})

duracion_3 = pd.DataFrame({"Numero" : [1,2,3,4,7,8],
                         "Duracion (min)" : [136, 142, 140, 121, 138, 152],
                           "Titulo" : ["Episodio I", "Episodio II", 
                                      "Episodio III", "Episodio IV",
                                      "Episodio VII", "Episodio VIII"]
                          })

print(titulos)
print("")
print(duracion)
print("")
print(duracion_2)
print("")
print(duracion_3)
```

```python
peliculas = titulos.merge(duracion, how="left", on="Episodio")
display(peliculas)
```

```python
peliculas = titulos.merge(duracion, how="right", on="Episodio")
display(peliculas.sort_values(by="Episodio"))
```

```python
peliculas = titulos.merge(duracion_2, how="inner", left_on="Episodio", right_on="Numero")
display(peliculas.sort_values(by="Episodio"))
```

```python
peliculas = titulos.merge(duracion_3, how="outer", left_on="Episodio", right_on="Numero",
                          suffixes=("_left", "_right"))
display(peliculas.sort_values(by="Episodio"))
```

## Librerías

Las librerías utilizadas en este documento son:

```python
# Tratamiento de datos
# ==============================================================================
import numpy as np
import pandas as pd
from greykite.framework.input.univariate_time_series import UnivariateTimeSeries

# Gráficos
# ==============================================================================
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
%matplotlib inline
import plotly
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True) 

# Modelado y Forecasting
# ==============================================================================
from greykite.common.data_loader import DataLoader
from greykite.framework.templates.autogen.forecast_config import ForecastConfig
from greykite.framework.templates.autogen.forecast_config import MetadataParam
from greykite.framework.templates.forecaster import Forecaster
from greykite.framework.templates.model_templates import ModelTemplateEnum
from greykite.framework.utils.result_summary import summarize_grid_search_results

# Configuración warnings
# ==============================================================================
import warnings
warnings.filterwarnings('ignore')
```

## Datos

In this example, we’ll load a dataset representing log(daily page views) on the Wikipedia page for Peyton Manning. It contains values from 2007-12-10 to 2016-01-20.

```python
# Descarga de datos
# ==============================================================================
url = ('https://raw.githubusercontent.com/JoaquinAmatRodrigo/'
       'Estadistica-machine-learning-python/master/data/visitas_wikipedia.csv')
datos = pd.read_csv(url, sep=',')
datos.head(3)
```

```python
# Prepraración del dato
# ==============================================================================
# datos['date'] = pd.to_datetime(datos['date'], format='%Y-%m-%d')
# datos = datos.set_index('date', drop=False)
# datos = datos.asfreq('1D')
# datos = datos.sort_index()
# datos[datos.y.isna()].shape
```

```python
# Almacenar el dato en un UnivariateTimeSeries de greykite
# ==============================================================================
ts = UnivariateTimeSeries()
ts.load_data(
    df          =  datos,
    time_col    = 'date',
    value_col   = 'y',
    freq        = 'D',
    date_format = '%Y-%m-%d'
)
```

```python
# Información del indice temporal y de los valores de la serie
# ==============================================================================
print('Información de las fechas')
print('=========================')
print(ts.time_stats)
print("")
print('Información de los valores')
print('=========================')
print(ts.value_stats)
```

La serie temporal tiene 59 fechas vacias. Los modelos de **greykite** interpolan automaticamente los valorea usentes:
    
    Greykite handles missing values and even missing timestamps. You should not impute the missing values on your own.

Greykite’s imputation methods prevent leakage of future information into the past during time-series cross validation and backtesting. They allow imputation based on future values, but only within each training set

## Gráficos

```python
fig = ts.plot(title = 'Serie temporal log(visitas página)')
plotly.io.show(fig)
```

```python
# Separación datos train-validation-test
# ==============================================================================
dias_test = 100
fecha_fin_train = datos.index[-1] - pd.Timedelta(dias_test, unit='D')
fecha_inicio_validacion = fecha_fin_train + pd.Timedelta(1, unit='H')
fecha_fin_validacion = fecha_inicio_validacion + pd.Timedelta(dias_validacion, unit='D')
fecha_inicio_test = fecha_fin_validacion + pd.Timedelta(1, unit='H')

datos_train = datos.loc[: fecha_fin_train, :]
datos_val   = datos.loc[fecha_inicio_validacion:fecha_fin_validacion, :]
datos_test  = datos.loc[fecha_inicio_test:, :]
```

```python
# Gráfico serie temporal
# ==============================================================================
fig, ax = plt.subplots(figsize=(10, 4))
datos_train.solar.plot(ax=ax, label='train', linewidth=0.5)
datos_val.solar.plot(ax=ax, label='validación', linewidth=0.5)
datos_test.solar.plot(ax=ax, label='test', linewidth=0.5)
ax.set_title('Producción energía solar')
ax.legend();
```

## Modelado

```python
# Definición metadatos del conjunto de datos
# ==============================================================================
metadata = MetadataParam(
 time_col  = "ds",# nombre de la columna con la fecha
 value_col = "y", # nombre de la columna con el valor de la serie
 freq      = "D"  # Frecuencia de la serie
)
```

```python
# Definición y entrenamiento del forecaster
# ==============================================================================
forecaster = Forecaster()  # Creates forecasts and stores the result
result = forecaster.run_forecast_config(  # result is also stored as `forecaster.forecast_result`.
             df = datos,
             config = ForecastConfig(
                         model_template=ModelTemplateEnum.SILVERKITE.name,
                         forecast_horizon=365,  # forecasts 365 steps ahead
                         coverage=0.95,         # 95% prediction intervals
                         metadata_param=metadata
                     )
            )
```

The output of run_forecast_config is a dictionary that contains the future forecast, historical forecast performance, and the original timeseries.

```python
backtest = result.backtest
fig = backtest.plot()
plotly.io.show(fig)
```

```python
result.backtest.df_train
```

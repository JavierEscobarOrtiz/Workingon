```python
from sklearn.datasets import fetch_openml
import pandas as pd
import numpy as np

bike_sharing = fetch_openml("Bike_Sharing_Demand", version=3, as_frame=True)
datos = bike_sharing.frame
datos.head(3)
```

```python
datos['year'] = datos['year'] + 2011
datos['year'] = datos['year'].astype(int).astype(str)
datos['month'] = datos['month'].astype(int).astype(str).apply(lambda x: x.zfill(2))
datos['hour'] = datos['hour'].astype(int).astype(str)
datos['weekday'] = datos['weekday'].astype(int).astype(str)

datos['date_time'] = datos[['year', 'month', 'weekday', 'hour']].agg('-'.join, axis=1)
datos['date_time'] = pd.to_datetime(datos['date_time'], format="%Y-%m-%w-%H")
datos = datos.set_index('date_time')
```

```python
datos
```

```python
# Verificar que un índice temporal está completo
# ==============================================================================
len(datos.index) == len(pd.date_range(start=datos.index.min(),
                                     end=datos.index.max(),
                                     freq='1h'))
```

```python
# Completar huecos en un índice temporal
# ==============================================================================
datos.asfreq(freq='30min', fill_value=np.nan)
```

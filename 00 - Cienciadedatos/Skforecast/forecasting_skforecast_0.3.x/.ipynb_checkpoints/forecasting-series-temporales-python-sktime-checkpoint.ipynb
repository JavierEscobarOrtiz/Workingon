## Introducción
<br>

Una [serie temporal](https://es.wikipedia.org/wiki/Serie_temporal) (*time series*) es una sucesión de datos ordenados cronológicamente, espaciados a intervalos iguales o desiguales. 

El término *forecasting* hace referencia al proceso de predecir el valor futuro de una serie temporal. Uno de los métodos de *forcasting* más utilizado es el autoregresivo, en el que el valor a futuro de una serie se modela en función del comportamiento pasado de dicha serie.

A lo largo de este documento, se describen de forma progresiva las principales funcionalidades que ofrece **python** para realizar predicciones sobre series temporales.
<br><br>

## Forecasting
<br>

Cunado se trabaja con series temporales, raramente si necesita predecir el siguiente elemento de la serie ($t+1$), sino todo un intervalo futuro o un punto alejado en el tiempo ($t+n$). 

Dado que, para predecir el momento $t+2$ se necesita el valor de $t+1$, y $t+1$ se desconoce, se necesita hacer predicciones recursivas en las que cada nueva predicción se basa en la predicción anterior. A este proceso se le conoce *recursive forecasting* o *multi-step forecasting* y es la principal diferencia a respecto a los problemas de regresión convencionales.


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1234)
y = 10 * np.sin(np.arange(0, 10, 0.1))
y = y + np.random.normal(loc=0.0, scale=4, size = len(y))
datetime = pd.date_range(start='1/1/2020', periods=len(y), freq='D')
datos = pd.Series(y, index=datetime)

fig, ax = plt.subplots(figsize=(9, 4))
datos[:-30].plot(ax=ax, c='c', marker='o', label='pasado',  zorder=3)
datos[-31:].plot(ax=ax, c='gray', label='futuro')
datos.iloc[[-30]].plot(ax=ax, c='red', marker='o', linestyle='', markersize=10,
                       label='valor a predecir')
ax.set_title('Predecir siguiente punto en el tiempo')
ax.legend();
```

## Forecasting sklearn

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

### Funciones auxiliares

```python
# Funciones auxiliares
# ==============================================================================

def transform_ts_to_matrix(serie, window_size, return_index=False, verbose=False):
    '''
    Esta función combierte una serie en una matriz de ventanas temporales
    donde, cada valor de y, está asociado a la ventana temporal de tamaño
    `window_size` que le precede.
    '''
    
    # Número de divisiones (ventanas)
    n_splits = len(serie) - window_size
    
    # Matriz donde almacenar los índices que forman cada ventana de datos
    X_index = np.full(shape=(n_splits, window_size), fill_value=np.nan, dtype=serie.index.dtype)
    y_index = np.full(shape=(n_splits, 1) , fill_value=np.nan, dtype=serie.index.dtype)
    
    # Matriz donde almacenar los datos que forman parte de cada ventana
    X_data  = np.full(shape=(n_splits, window_size), fill_value=np.nan, dtype= float)
    y_data  = np.full(shape=(n_splits, 1), fill_value=np.nan, dtype= float)
    
    for i in range(n_splits):
        
        train_index = np.array(serie.index[i : window_size + i])
        test_index  = serie.index[window_size + i]
        
        if verbose:
            print("TRAIN:", train_index, "TEST:", test_index)
            
        X_index[i, :] = train_index
        y_index[i]    = test_index

        X_data[i, :] = serie[train_index]
        y_data[i]    = serie[test_index]
    
    if return_index:
        return X_index, y_index
    else:
        return X_data, y_data
    

def ts_train_test_split(serie, test_size=2, window_size=10, return_index=False, verbose=False):
    '''
    Esta función combierte una serie en una matriz de ventanas temporales
    donde, cada valor de y, está asociado a la ventana temporal de tamaño
    `window_size` que le precede. Después, la divide en dos partes (train y test).
    '''
    X, y = transform_ts_to_matrix(
                serie        = serie,
                window_size  = window_size,
                return_index = return_index,
                verbose      = verbose
            )
    X_train = X[:-test_size, :]
    y_train = y[:-test_size]
    X_test  = X[-test_size, :].reshape(1, -1)
    y_test  = y[-test_size:]
    
    return X_train, X_test, y_train, y_test


def forecaster_predict(modelo, X_test, steps):
    '''
    Proceso de predicción recursiva en el que cada predicción se utiliza como
    predictor de la siguiente predicción.
    '''

    predicciones = predicciones = np.full(shape=steps, fill_value=np.nan)

    for i in range(steps):

        prediccion = modelo.predict(X_test).flatten()[0]
        predicciones[i] = prediccion

        # Actualizar valores de X_test
        X_test = X_test.flatten()
        X_test = np.append(X_test[1:], prediccion)
        X_test = X_test.reshape(1, -1)


    return(predicciones)

def extender_indice(index, steps):
    '''
    Extender índice temporal con n elementos, manteniendo la frecuencia del
    índice.
    '''
    new_index = index.union([index[-1] + steps*datos.index.freq])
    
    return new_index
```

#### Ejemplos funcionamiento

```python
set_prueba = pd.Series(range(21))
X, y = transform_ts_to_matrix(serie = set_prueba, window_size=10, verbose=True)

X_train, X_test, y_train, y_test = ts_train_test_split(
                                        serie        = set_prueba,
                                        test_size    = 2,
                                        window_size  = 10,
                                        return_index = True
                                    )
print('X_train')
print(X_train)
print("")
print('X_test')
print(X_test)
print("")
print('y_train')
print(y_train)
print("")
print('y_test')
print(y_test)
```

```python
# Preprocesado del dato
# ==============================================================================
X_train, X_test, y_train, y_test = ts_train_test_split(
                                        serie        = datos,
                                        test_size    = 36,
                                        window_size  = 12,
                                        return_index = False
                                    )
```

```python
# Entrenamiento del modelo
# ==============================================================================
modelo = RandomForestRegressor(random_state=123)
modelo.fit(X=X_train, y=y_train)
```

```python
# Predicciones
# ==============================================================================
steps = 36
predicciones = forecaster_predict(modelo=modelo, X_test=X_test, steps=steps)
# Se añade el índice a las predicciones
predicciones = pd.Series(data=predicciones, index=datos.index[-steps :])
predicciones.head()
```

```python
fig, ax=plt.subplots(figsize=(9, 4))
datos[:-steps].plot(ax=ax, label='train')
datos[-steps:].plot(ax=ax, label='test')
predicciones.plot(ax=ax, label='autoregresivo sklearn')
ax.legend();
```

```python
# Error
# ==============================================================================
mse_autoreg_sklearn = mean_squared_error(
                            y_true = datos[-steps:],
                            y_pred = predicciones
                        )
mse_autoreg_sklearn
```

## Forecasting statsmodels
<br>

### Autoregresivo

```python
from statsmodels.tsa import ar_model
```

```python
horizonte_prediccion = 36
X_train = datos[:-horizonte_prediccion]
X_test  = datos[-horizonte_prediccion:]
```

```python
# Modelo autorregresivo
# ==============================================================================
modelo_ar = ar_model.AR(X_train)
res = modelo_ar.fit(maxlag=12)
```

```python
# Predicciones
# ==============================================================================
predicciones_ar = res.predict(start=X_test.index[0], end=X_test.index[-1])
predicciones_ar.head()
```

```python
fig, ax=plt.subplots(figsize=(9, 4))
X_train.plot(ax=ax, label='train')
X_test.plot(ax=ax, label='test')
predicciones_ar.plot(ax=ax, label='autoregresivo')
ax.legend();
```

```python
# Error
#===============================================================================
mse_autoreg_stast = mean_squared_error(
                        y_true = X_test,
                        y_pred = predicciones_ar
                    )
mse_autoreg_stast
```

### ARIMA

```python
from statsmodels.tsa.arima import model
```

```python
# Modelo arima
# ==============================================================================
arima_model = model.ARIMA(X_train, order=(12, 1, 0))
res = arima_model.fit()
```

```python
# Predicciones
# ==============================================================================
predicciones_arima = res.predict(start=X_test.index[0], end=X_test.index[-1])
predicciones_arima.head()
```

```python
# Error
#===============================================================================
mse_arima_stast = mean_squared_error(
                        y_true = X_test,
                        y_pred = predicciones_arima
                    )
mse_arima_stast
```

```python
fig, ax=plt.subplots(figsize=(9, 4))
X_train.plot(ax=ax, label='train')
X_test.plot(ax=ax, label='test')
predicciones_arima.plot(ax=ax, label='arima')
ax.legend();
```

## SKtime

https://www.sktime.org/en/latest/examples/01_forecasting.html

### Sktime con modelos scikit-learn
<br>

Con la clase `ReducedRegressionForecaster` se puede convertir un modelo cualquiera de scikitlearn en un forecaster.

```python
from sktime.forecasting.model_selection import temporal_train_test_split
from sktime.forecasting.model_selection import ForecastingGridSearchCV
from sktime.forecasting.model_selection import SlidingWindowSplitter
from sktime.forecasting.compose import ReducedRegressionForecaster
from sktime.forecasting.arima import ARIMA, AutoARIMA
from sktime.forecasting.exp_smoothing import ExponentialSmoothing
from sktime.forecasting.base import ForecastingHorizon


from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
```

```python
y_train, y_test = temporal_train_test_split(datos, test_size=36)
```

```python
# Modelo autoregresivo
# ==============================================================================
regressor = RandomForestRegressor(random_state=123)
forecaster = ReducedRegressionForecaster(
                regressor     = regressor,
                window_length = 12,
                strategy      = "recursive"
             )
forecaster.fit(y_train)
```

```python
# Predicciones
# ==============================================================================
predicciones = forecaster.predict(fh=np.arange(1,37))
predicciones.head()
```

```python
# Error
#===============================================================================
mse_sktime_scikit = mean_squared_error(
                        y_true = y_test,
                        y_pred = predicciones
                    )
mse_sktime_scikit
```

```python
fig, ax=plt.subplots(figsize=(9, 4))
y_train.plot(ax=ax, label='train')
y_test.plot(ax=ax, label='test')
predicciones.plot(ax=ax, label='sktime random forest')
ax.legend();
```

### Sktime + GridSearch

```python
# Hiperparámetros
# ==============================================================================
regressor_param_grid = {"n_estimators": [100, 200, 300]}
forecaster_param_grid = {"window_length": [5, 15, 20, 25]}

# Se crea el grid search del regressor de scikitlearn
regressor = GridSearchCV(
                estimator  = RandomForestRegressor(),
                param_grid = regressor_param_grid,
                scoring    = 'neg_root_mean_squared_error'
            )

# Se crea el forecaster
forecaster = ReducedRegressionForecaster(
                regressor     = regressor,
                window_length = 15, # Este valor será sobreescrito con el grid
                strategy      = "recursive"
             )

# Se crea el gridsearch del forecaster con un cv temporal con la mitad de los 
# datos de entrenamiento para la iteración de partida y bloques de 36 predicciones
cv = SlidingWindowSplitter(initial_window=int(len(y_train) * 0.5), fh=36)
gscv = ForecastingGridSearchCV(
         forecaster = forecaster,
         cv         = cv,
         param_grid = forecaster_param_grid
       )

_ = gscv.fit(y_train)
```

```python
# Resultados Grid
# ==============================================================================
pd.DataFrame(gscv.cv_results_)
```

```python
# Mejores hiperparámetros encontrados
# ==============================================================================
print(gscv.best_params_, gscv.best_forecaster_.regressor_.best_params_)
```

```python
# Modelo final
# ==============================================================================
regressor = RandomForestRegressor(random_state=123, **gscv.best_forecaster_.regressor_.best_params_)
forecaster = ReducedRegressionForecaster(
                regressor     = regressor,
                window_length = gscv.best_params_['window_length'],
                strategy      = "recursive"
             )
_ = forecaster.fit(y_train)
```

```python
# Predicciones
# ==============================================================================
predicciones = forecaster.predict(fh=np.arange(1,37))
predicciones.head()
```

```python
# Error
#===============================================================================
mse_sktime_scikit = mean_squared_error(
                        y_true = y_test,
                        y_pred = predicciones
                    )
mse_sktime_scikit
```

```python
fig, ax=plt.subplots(figsize=(9, 4))
y_train.plot(ax=ax, label='train')
y_test.plot(ax=ax, label='test')
predicciones.plot(ax=ax, label='sktime random forest')
ax.legend();
```

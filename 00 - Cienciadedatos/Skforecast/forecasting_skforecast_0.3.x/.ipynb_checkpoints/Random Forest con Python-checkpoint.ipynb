## Ejemplo

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_graphviz
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(
    cancer.data,
    cancer.target,
    stratify = cancer.target,
    random_state = 123
)
```

```python
modelo_tree = DecisionTreeClassifier(max_depth=2)
modelo_tree.fit(X_train, y_train)
```

```python
fig, ax = plt.subplots(figsize=(7, 5))

plot = plot_tree(
            decision_tree = modelo_tree,
            feature_names = cancer.feature_names,
            class_names   = cancer.target_names,
            filled        = True,
            ax = ax
       )
```

```python
# No pruning
modelo_tree = DecisionTreeClassifier()
modelo_tree.fit(X_train, y_train)
fig, ax = plt.subplots(figsize=(8, 5))

plot = plot_tree(
            decision_tree = modelo_tree,
            feature_names = cancer.feature_names,
            class_names   = cancer.target_names,
            filled        = True,
            ax = ax
       )
```

```python
# Early stoping (pre-pruning)

modelo_tree = DecisionTreeClassifier(max_depth=4, min_samples_leaf=20)
modelo_tree.fit(X_train, y_train)
fig, ax = plt.subplots(figsize=(8, 5))

plot = plot_tree(
            decision_tree = modelo_tree,
            feature_names = cancer.feature_names,
            class_names   = cancer.target_names,
            filled        = True,
            ax = ax
       )
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'max_depth':range(1, 7)}
grid = GridSearchCV(DecisionTreeClassifier(random_state=0),
                    param_grid=param_grid,
                    cv=10,
                    return_train_score=True)
grid.fit(X_train, y_train)
```

```python
fig, ax = plt.subplots(figsize=(7, 5))

scores = pd.DataFrame(grid.cv_results_)
scores.plot(x='param_max_depth', y='mean_train_score', yerr='std_train_score', ax=ax)
scores.plot(x='param_max_depth', y='mean_test_score', yerr='std_test_score', ax=ax);
```

```python
# Post pruning (const complexity pruning)
# =============================================================================
#Primero se crece al máximo el árbol y despues se poda 
modelo_tree = DecisionTreeClassifier(ccp_alpha=0.05)
modelo_tree.fit(X_train, y_train)
fig, ax = plt.subplots(figsize=(8, 5))

plot = plot_tree(
            decision_tree = modelo_tree,
            feature_names = cancer.feature_names,
            class_names   = cancer.target_names,
            filled        = True,
            ax = ax
       )
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'ccp_alpha':np.linspace(0, 0.1, 20)}
grid = GridSearchCV(DecisionTreeClassifier(random_state=0),
                    param_grid=param_grid,
                    cv=10,
                    return_train_score=True)
grid.fit(X_train, y_train)

fig, ax = plt.subplots(figsize=(7, 5))

scores = pd.DataFrame(grid.cv_results_)
scores.plot(x='param_ccp_alpha', y='mean_train_score', yerr='std_train_score', ax=ax)
scores.plot(x='param_ccp_alpha', y='mean_test_score', yerr='std_test_score', ax=ax);
```

Importante!!:Pruning puede general mejores resultados ya que puede, permite que una división poco útil se mantenga si siguientes dicisiones sí  lo son.

## Extrapolación con árboles

Los árboles de regresión no extrapolan fuera del rángo de entrenamiento. Para valores fuera del rango, la predicción siempre es la media del nodo más cercano.

```python
X_train = np.linspace(0, 100, 50)
y_train = (X_train + 100) + np.random.normal(loc=0.0, scale=5.0, size=X_train.shape)

X_test = np.linspace(101, 150, 50)
y_test = (X_test + 100) + np.random.normal(loc=0.0, scale=5.0, size=X_test.shape)
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(X_train, y_train, c='black')
ax.plot(X_test, y_test, c='blue')
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression

modelo_lineal = LinearRegression()
modelo_lineal.fit(X_train.reshape(-1, 1), y_train)

modelo_tree = DecisionTreeRegressor()
modelo_tree.fit(X_train.reshape(-1, 1), y_train)


prediccion_lineal = modelo_lineal.predict(np.concatenate([y_train, y_test]).reshape(-1, 1)).flatten()
prediccion_tree   = modelo_tree.predict(np.concatenate([y_train, y_test]).reshape(-1, 1)).flatten()

fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(X_train, y_train, c='black')
ax.plot(X_test, y_test, c='blue')
ax.plot(np.concatenate([y_train, y_test]), prediccion_lineal, c='green')
ax.plot(np.concatenate([y_train, y_test]), prediccion_tree, c='orange')
```
